```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant = 0, n_classes = 2, n_clusters_per_class = 1,random_state=41, hypercube = False, class_sep = 15)
```

```python
print(X.shape, y.shape)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap = 'winter',s = 100)
plt.show()
```

```python
def perceptron(X,y):

  w1 = w2 = b = 1 #Initializing var with random values = 1
  lr = 0.1

  for j in range(1000):
    for i in range(X.shape[0]):
      z = w1*X[i][0] + w2*X[i][1] + b     # z = f(xi)
      if z*y[i] < 0:
        w1 = w1 + lr*X[i][0]*y[i]
        w2 = w2 + lr*X[i][1]*y[i]
        b = b + lr*y[i]

  return w1,w2,b


```

```python
w1,w2,b = perceptron(X,y)
```

```python
print('w1 = ',w1)
print('w2 = ',w2)
print('b = ',b)
```

# **Visualization**

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c

plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap = 'winter',s = 100)
plt.plot(x_input,y_input,color = 'red')
plt.show()
```
